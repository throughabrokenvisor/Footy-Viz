##Treemaps for Assists at Lille OSC/2019-20


```python
data = pd.read_csv("Lille_Stats.csv")
```

```python
data
```

```python
data.columns = ['Player','Nation','Position','Age','MP','Starts','Min','90s','Gls','Ast','G-PK','PK','PKatt','CrdY','CrdR','Gls90','Ast90','G+A','G-PK','G+A-PK','xG','npxG','xA','npxG+xA','xG90','xA90','xG+xA90','npxG','npxG+xA','Matches']
```

```python
data1 = data.drop(data.index[0])
```

```python
data1
```

```python
data1["Ast"] = data1.Ast.astype('int')
```

```python
dataAssists = data1[data1["Ast"]>1]
dataAssists
```

```python
norm = matplotlib.colors.Normalize(vmin=min(dataAssists.Ast), vmax=max(dataAssists.Ast))
colors = [matplotlib.cm.Blues(norm(value)) for value in dataAssists.Ast]
```

```python
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(18, 6)
```

```python
squarify.plot(label=dataAssists.Player,sizes=dataAssists.Ast, color = colors, alpha=.6, pad = True)
plt.title("Lille Assists",fontsize=21,fontweight="bold")
plt.axis('off')
plt.show()
```

###Radar Charts



```python
createRadar("Osimhen at Lille/2019-2020",[13,82,32,39.0,13.5])
```

```python
createRadar("Osimhen at SSC Napoli/2020-2021",[10,64,26,40.6,13.6])
```

```python
createRadar("Osimhen at SSC Napoli/2021-2022",[5,39,10,25.6,13.8])
```

```python
def createRadar(player, data):
    Attributes = ["Goals","Shots","SoT","SoT%","Dist"]
    
    data += data [:1]
    
    angles = [n / 5 * 2 * pi for n in range(5)]
    angles += angles [:1]
    
    ax = plt.subplot(111, polar=True)

    plt.xticks(angles[:-1],Attributes)
    ax.plot(angles,data)
    ax.fill(angles, data, 'blue', alpha=0.1)

    ax.set_title(player)
    plt.show()
```
