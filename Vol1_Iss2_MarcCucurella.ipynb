##Radar Charts for analyzing the contributions of various players while in possession

```python
def createRadar(Touches, data):
    Attributes = ["Marc Cucurella","Joel Veltman","Lewis Dunk","Adam Lallana"]
    
    data += data [:1]
    
    angles = [n / 4 * 2 * pi for n in range(4)]
    angles += angles [:1]
    
    ax = plt.subplot(111, polar=True)

    plt.xticks(angles[:-1],Attributes)
    ax.plot(angles,data)
    ax.fill(angles, data, 'blue', alpha=0.2)

    ax.set_title(Touches)
    plt.show()
```

```python
createRadar("Players with most touches(Top 4)",[1413,1090,1122,966])
```
